```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

Reading in dataset:

```python
url = 'https://raw.githubusercontent.com/thaonguyyen/DS2002-HW/main/Top_Highest_Openings.csv'
df = pd.read_csv(url)
df.head()
```

Grouped bar chart:

```python
grouped_data = df.groupby('Distributor').agg({'Opening': 'sum', 'Total Gross': 'sum'}).reset_index()
labels = grouped_data['Distributor']
opening_means = grouped_data['Opening']
gross_means = grouped_data['Total Gross']

x = np.arange(len(labels))  # the label locations
width = 0.5  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, opening_means, width, label='Openings')
rects2 = ax.bar(x + width/2, gross_means, width, label='Total Gross')

ax.set_ylabel('Dollars')
ax.set_xlabel('Distributors')
ax.set_title('Distributors by Opening and Total Gross')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend();

plt.xticks(rotation=90)  # rotate labels to prevent overlap
plt.show()
```

Scatter plot:

```python
plt.scatter(df['Opening'], df['Total Gross'])
plt.title('Opening vs Total Gross')
plt.xlabel('Opening Earnings')
plt.ylabel('Total Gross Earnings')
plt.grid(True)
plt.show()
```

Box plot:

```python
grouped = df.groupby('Distributor')['Total Gross'].apply(list)
data = [values for values in grouped]
labels = grouped.index

fig, ax = plt.subplots()
ax.boxplot(data, labels=labels, vert=False)
ax.set_title('Total Gross by Distributor')
ax.set_xlabel('Total Gross')
ax.set_ylabel('Distributor')
plt.show()
```
